```python
pages=input('Enter the range of pages to be scraped ( Ex: 1,20) : ')

urls=[]
```

```python
data=[]
for url in urls:
    try:
        print(url)
        driver.get(url)
        time.sleep(random.randint(3,5))

        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(random.randint(2,4))
        project_block=driver.find_element_by_css_selector('section[class="js-project-content js-project-description-content project-content"]')
        project_text=driver.execute_script("return Array.from(document.querySelectorAll(\
                                           #'content-wrap > div.NS_projects__content > section.js-project-content.js-project-description-content.project-content > div > div > div > div.col.col-8.description-container'))\
                                          # .map(function(element) {return element.textContent;})")

        no_of_images=len(project_block.find_elements_by_tag_name('img'))
        no_of_videos=len(project_block.find_elements_by_tag_name('video'))

        try:
            first_video = project_block.find_element_by_css_selector('div[class="project-image mb2 cf aspect-ratio--object"]').get_attribute('data-has-video')
        except:
            first_video='False'

        name=driver.find_element_by_css_selector('a[class="hero__link"]').text
        backers=project_block.find_elements_by_css_selector('h3[class="mb0"]')[1].text
        backers=float(backers.replace(',',''))
        category=project_block.find_elements_by_css_selector('a[class="grey-dark mr3 nowrap type-12"]')[1].text
        country = project_block.find_elements_by_css_selector('a[class="grey-dark mr3 nowrap type-12"]')[0].text
        funded=project_block.find_elements_by_css_selector('span[class="money"]')[0].text
        funded=funded.replace(',','').replace(' ','')
        currency,funded=currency_finder(funded)
        funded=float(funded)
        goal=project_block.find_elements_by_css_selector('span[class="money"]')[1].text
        goal=goal.replace(',','').replace(' ','')
        currency,goal=currency_finder(goal)
        goal=float(goal)
        percentage_funded = round((funded/goal))
        percentage_funded=float(percentage_funded)
        duration_tag=project_block.find_element_by_css_selector('p[class="f5"]')
        end_date= duration_tag.find_elements_by_css_selector('time[class="js-adjust-time"]')[1].get_attribute('datetime').split('T')[0]
        start_year = float(duration_tag.find_elements_by_css_selector('time[class="js-adjust-time"]')[0].get_attribute('datetime').split('T')[0].split('-')[0])
        end_year = float(duration_tag.find_elements_by_css_selector('time[class="js-adjust-time"]')[1].get_attribute('datetime').split('T')[0].split('-')[0])
        duration=duration_tag.text.split('(')[1].split(' ')[0]
        duration = float(duration)

        second_pledge=project_block.find_elements_by_css_selector('div[class="pledge__info"]')[1]
        delivery=second_pledge.find_element_by_css_selector('span[class="pledge__detail-info"]').text
        delivery_date=second_pledge.find_element_by_tag_name('time').get_attribute('datetime').split('T')[0]

        start_delivery = datetime.strptime(end_date,'%Y-%m-%d')-datetime.strptime(delivery_date,'%Y-%m-%d')
        start_delivery = abs(start_delivery.days)
        total_duration = duration + start_delivery

        level_pledges=len(project_block.find_elements_by_css_selector('div[class="pledge__info"]'))
        level_pledges

        driver.get(url.split('?')[0]+'/community')
        time.sleep(2)

        new_backers=driver.find_element_by_xpath('//div[@class="new-backers"]/div[@class="count"]').text
        new_backers=float(new_backers.replace(',',''))
        existing__backers=driver.find_element_by_xpath('//div[@class="existing-backers"]/div[@class="count"]').text
        existing__backers=float(existing__backers.replace(',',''))

        update=driver.find_element_by_xpath('//a[@data-content="updates"]/span').text
        update=float(update.replace(',',''))

        driver.find_element_by_css_selector('a[data-modal-class="modal_project_by"]').click()
        time.sleep(2)
        #entrepreneur = driver.find_element_by_css_selector('h1[class="f2 normal mb2"]').text
        projects=driver.find_element_by_css_selector('div[class="created-projects py2 f5 mb3"]').text
        projects_launched=projects.split(' · ')[0].split(' ')[0].strip()
        if projects_launched=='First':
            projects_launched=0
        else:
            projects_launched=float(projects_launched)
        projects_backed=float(projects.split(' · ')[1].split(' ')[0].strip())


        #print('Category ',category)
        #print('No.of of backers',backers)
        #print('No.of images and videos',no_of_images,no_of_videos)
        #word_count=len(project_text[0].split(' '))
        #print(word_count)
        #print('Funded: ', funded.replace('$',''))
        #print('Goal : ', goal.replace('$',''))

        data.append([name,category, country, backers, funded, goal, percentage_funded,
                     duration, start_delivery, total_duration,
                     level_pledges, new_backers,existing__backers, update, start_year,
                     end_year, currency, projects_launched, projects_backed, first_video])

    except Exception as e:
        print(e)
        print('failed for this url : ', url)
```
